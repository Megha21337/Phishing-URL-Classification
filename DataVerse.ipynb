```python
# URL Length Hypothesis:

import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

dataset = pd.read_csv('PhiUSIIL_Phishing_URL_Dataset.csv')

phishing_urls = dataset[dataset['label'] == 1]
legitimate_urls = dataset[dataset['label'] == 0]

phishing_url_length = phishing_urls['URLLength']
legitimate_url_length = legitimate_urls['URLLength']

t_stat, p_value = ttest_ind(phishing_url_length, legitimate_url_length, equal_var=False)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

plt.figure(figsize=(12, 6))
plt.hist(phishing_url_length, bins=50, alpha=0.5, label='Phishing URLs', color='red')
plt.hist(legitimate_url_length, bins=50, alpha=0.5, label='Legitimate URLs', color='green')
plt.xlabel('URL Length')
plt.ylabel('Frequency')
plt.title('Distribution of URL Lengths for Phishing vs Legitimate URLs')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# HTTPS Usage Hypothesis:

import pandas as pd
from scipy.stats import chi2_contingency

dataset = pd.read_csv('PhiUSIIL_Phishing_URL_Dataset.csv')

phishing_urls = dataset[dataset['label'] == 1]
legitimate_urls = dataset[dataset['label'] == 0]

phishing_https = phishing_urls['IsHTTPS']
legitimate_https = legitimate_urls['IsHTTPS']

contingency_table = pd.crosstab(index=dataset['label'], columns=dataset['IsHTTPS'])

chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")



```

```python
# Visualize HTTPS usage differences between phishing and legitimate URLs using a bar chart

https_counts = contingency_table
plt.figure(figsize=(10, 6))
https_counts.plot(kind='bar', stacked=False, color=['red', 'green'], alpha=0.7)
plt.xlabel('URL Type (0 = Legitimate, 1 = Phishing)')
plt.ylabel('Count')
plt.title('HTTPS Usage in Phishing vs Legitimate URLs')
plt.xticks(rotation=0)
plt.legend(['No HTTPS', 'HTTPS'], title="HTTPS Usage")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Hypothesis Statement
# Null Hypothesis (Hâ‚€): There is no significant difference in the average number of subdomains between phishing URLs and legitimate URLs.
# Alternative Hypothesis (Hâ‚): The average number of subdomains is significantly different between phishing and legitimate URLs.

phishing_subdomains = phishing_urls['NoOfSubDomain']
legitimate_subdomains = legitimate_urls['NoOfSubDomain']

t_stat_subdomains, p_value_subdomains = ttest_ind(phishing_subdomains, legitimate_subdomains, equal_var=False)

subdomain_hypothesis_test_summary = {
    "t_statistic": t_stat_subdomains,
    "p_value": p_value_subdomains
}

subdomain_hypothesis_test_summary
```
